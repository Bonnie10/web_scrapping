```python
import numpy as np
import pandas as pd
import requests 
from bs4 import BeautifulSoup 
```

```python
url = 'https://books.toscrape.com/'
htmlPage = requests.get(url = url)
Soup = BeautifulSoup(htmlPage.content, 'html.parser')
```

```python
print(len(Soup.find_all('li',{"class":"col-xs-6 col-sm-4 col-md-3 col-lg-3"})))
Books = Soup.find_all('li',{"class":"col-xs-6 col-sm-4 col-md-3 col-lg-3"})
```

```python
def BookScrapping (book):
    infor = {}
    infor['Title'] = book.find('article',{'class':'product_pod'}).h3.a['title']
    infor['Url'] = book.find('article',{'class':'product_pod'}).h3.a['href']
    infor['Rating'] = book.find('p',{'class':regex})['class'][-1]
    infor['Price'] = book.find('div', {'class':'product_price'}).find('p',{'class':'price_color'}).text
    infor['Availability'] = book.find('div', {'class':'product_price'}).find('p',{'class':'instock availability'}).get_text(strip = True)
    return infor
```

```python
bookDict = [BookScrapping (book) for book in Books]
data = pd.DataFrame(bookDict)
data
```

```python
url = 'https://books.toscrape.com/catalogue/page-1.html'
```

```python
urls = ['https://books.toscrape.com/catalogue/page-{}.html'.format (i) for i in range (1,51)]
urls
```

```python
def get_20_books(url):
    htmlPage = requests.get(url)
    soup = BeautifulSoup(htmlPage.content,'html.parser')
    raw = soup.find_all('li',{'class':"col-xs-6 col-sm-4 col-md-3 col-lg-3"})
    to_dicts = [BookScrapping(book) for book in raw]
    return to_dicts
    
```

```python
all_dict = []
for url in urls:
    all_dict.extend(get_20_books(url))
print(len(all_dict))
all_dict
```

```python
data = pd.DataFrame(all_dict)
data
```
